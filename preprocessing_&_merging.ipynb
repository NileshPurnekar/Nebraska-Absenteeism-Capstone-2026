```python
import pandas as pd
import os

base_path = "/content/"

frl_files = {
    "2018-19": "2018-19_Free_and_Reduced_Lunch_Counts_by_School.xls",
    "2019-20": "2019-20_Free_and_Reduced_Lunch_Counts_by_School.xls",
    "2020-21": "2020-21_Free_and_Reduced_Lunch_Counts_by_School.xls",
    "2021-22": "2021-2022_Free_and_Reduced_Lunch_Counts_by_School.xlsx",
    "2022-23": "2022-2023_Free_and_Reduced_Lunch_Counts_by_School.xlsx",
    "2023-24": "2023-2024_Free_and_Reduced_Lunch_Counts_by_School.xlsx",
    "2024-25": "2024-2025_Free_and_Reduced_Lunch_Counts_by_School.xlsx",
}

mem_files = {
    "2018-19": "MembershipByGradeRaceAndGender_20182019.csv",
    "2019-20": "MembershipByGradeRaceAndGender_20192020.csv",
    "2020-21": "MembershipByGradeRaceAndGender_20202021.csv",
    "2021-22": "MembershipByGradeRaceAndGender_20212022.csv",
    "2022-23": "MembershipByGradeRaceAndGender_20222023.csv",
    "2023-24": "MembershipByGradeRaceAndGender_20232024.csv",
    "2024-25": "MembershipByGradeRaceAndGender_20242025.csv",
}

def normalize_id(x):
    return str(x).replace("-", "").strip().lstrip("0").zfill(8)

def load_frl(year):
    path = base_path + frl_files[year]
    xl = pd.ExcelFile(path)
    sheet = [s for s in xl.sheet_names if "School" in s][0]
    raw = pd.read_excel(path, sheet_name=sheet, header=None)
    header_row = raw[raw.apply(lambda r: r.astype(str).str.contains("CODISTSCH|SCHOOLNAME", case=False).any(), axis=1)].index[0]
    df = pd.read_excel(path, sheet_name=sheet, header=header_row)
    df.columns = df.columns.str.strip()
    frl_col = [c for c in df.columns if "FREEREDUCED" in c.upper().replace(" ", "")]
    df["FREEREDUCED"] = df[frl_col[0]] if frl_col else None
    return df

def load_mem(year):
    df = pd.read_csv(base_path + mem_files[year])
    df.columns = df.columns.str.strip()
    return df

all_years = []

for year in frl_files.keys():
    frl = load_frl(year)
    mem = load_mem(year)

    frl = frl[["CODISTSCH", "COUNTY", "SCHOOLNAME", "SCHOOLTYPE", "GRADE", "ENROLLED", "FREEREDUCED", "PERCENT"]].copy()
    frl.columns = ["SCHOOL_ID", "COUNTY", "SCHOOL_NAME", "SCHOOL_TYPE", "GRADE", "ENROLLED", "FRL_COUNT", "FRL_PERCENT"]
    frl = frl[frl["SCHOOL_ID"].notna()]

    mem_grouped = mem.groupby("CO_DIST_SCH")["TOTAL"].sum().reset_index()
    mem_grouped.columns = ["SCHOOL_ID", "TOTAL_STUDENTS"]

    frl["SCHOOL_ID"] = frl["SCHOOL_ID"].astype(str).apply(normalize_id)
    mem_grouped["SCHOOL_ID"] = mem_grouped["SCHOOL_ID"].astype(str).apply(normalize_id)

    merged = pd.merge(frl, mem_grouped, on="SCHOOL_ID", how="left")
    merged["YEAR"] = year
    all_years.append(merged)
    print(f"{year} done — {len(merged)} rows")

final_df = pd.concat(all_years, ignore_index=True)
final_df = final_df[["SCHOOL_ID", "YEAR", "COUNTY", "SCHOOL_NAME", "SCHOOL_TYPE", "GRADE", "FRL_PERCENT", "FRL_COUNT", "ENROLLED", "TOTAL_STUDENTS"]]

final_df.to_csv("/content/final_panel_dataset.csv", index=False)
print(f"\nDone! Shape: {final_df.shape}")
print(f"Unique schools: {final_df['SCHOOL_ID'].nunique()}")
final_df.head()
```

```python
absence_files = {
    "2018-19": "2018-2019-Student-Absences-Counts-by-District.xlsx",
    "2019-20": "2019-2020-Student-Absences-Counts-by-District.xlsx",
    "2020-21": "2020-2021-Student-Absences-Counts-by-District.xlsx",
    "2021-22": "2021-2022-Student-Absences-Counts-by-District.xlsx",
    "2022-23": "2022-2023-Student-Absences-Counts-by-District.xlsx",
    "2023-24": "2023-2024-Student-Absences-Counts-by-District.xlsx",
    "2024-25": "2024-2025-Student-Absences-Counts-by-District.xlsx",
}

def load_absence(year):
    path = "/content/" + absence_files[year]
    raw = pd.read_excel(path, sheet_name="DISTRICT TOTAL", header=None)
    header_row = raw[raw.apply(lambda r: r.astype(str).str.contains("DISTRICT CODE", case=False).any(), axis=1)].index[0]
    df = pd.read_excel(path, sheet_name="DISTRICT TOTAL", header=header_row)
    df.columns = df.columns.str.strip()
    df = df[df["DISTRICT CODE"].notna()]
    df = df[~df["DISTRICT CODE"].astype(str).str.contains("DISTRICT CODE", case=False)]

    chronic_col = [c for c in df.columns if "CHRONIC" in c.upper()]
    membership_col = [c for c in df.columns if c.strip().upper() == "MEMBERSHIP"]

    df["DISTRICT_ID"] = df["DISTRICT CODE"].astype(str).str.replace("-", "").str.strip()
    df["CHRONIC_ABSENT"] = pd.to_numeric(df[chronic_col[0]], errors="coerce")
    df["DISTRICT_MEMBERSHIP"] = pd.to_numeric(df[membership_col[0]], errors="coerce")
    df["CHRONIC_ABSENT_RATE"] = (df["CHRONIC_ABSENT"] / df["DISTRICT_MEMBERSHIP"]).round(4)
    df["YEAR"] = year

    return df[["DISTRICT_ID", "YEAR", "CHRONIC_ABSENT", "DISTRICT_MEMBERSHIP", "CHRONIC_ABSENT_RATE"]]

all_absence = []
for year in absence_files.keys():
    df = load_absence(year)
    all_absence.append(df)
    print(f"{year} — {len(df)} districts")

absence_panel = pd.concat(all_absence, ignore_index=True)
print(absence_panel.head())
```

```python
final_df = pd.read_csv("/content/final_panel_dataset.csv")

final_df["DISTRICT_ID"] = final_df["SCHOOL_ID"].astype(str).str[:8].str.strip()
absence_panel["DISTRICT_ID"] = absence_panel["DISTRICT_ID"].astype(str).str.strip()

ml_df = pd.merge(final_df, absence_panel[["DISTRICT_ID", "YEAR", "CHRONIC_ABSENT_RATE"]],
                 on=["DISTRICT_ID", "YEAR"], how="left")

print(f"Shape: {ml_df.shape}")
print(f"Missing target: {ml_df['CHRONIC_ABSENT_RATE'].isna().sum()}")
print(ml_df[["SCHOOL_ID", "YEAR", "FRL_PERCENT", "CHRONIC_ABSENT_RATE"]].head(10))
```

```python
print("Panel DISTRICT_ID sample:")
print(final_df["DISTRICT_ID"].head(10).tolist())

print("\nAbsence DISTRICT_ID sample:")
print(absence_panel["DISTRICT_ID"].head(10).tolist())
```

```python
# Absence: "010003000" → take chars 2-7 → "10003" → but panel has "10003001"
# Panel SCHOOL_ID "10003001" → first 5 chars = "10003" = district
# Absence DISTRICT_ID "010003000" → chars 1-6 = "100030" hmm

# Let's normalize both to just the 5-digit district code
final_df["DIST_KEY"] = final_df["SCHOOL_ID"].astype(str).str[1:6]
absence_panel["DIST_KEY"] = absence_panel["DISTRICT_ID"].astype(str).str[1:6]

print("Panel DIST_KEY:", final_df["DIST_KEY"].head(5).tolist())
print("Absence DIST_KEY:", absence_panel["DIST_KEY"].head(5).tolist())
```

```python
# Panel SCHOOL_ID "10003001" → strip leading 1 → "0003001" → first 5 = "00030"
# Absence DISTRICT_ID "010003000" → first 5 = "01000" nope
# Let's just strip ALL leading zeros and compare raw numbers

final_df["DIST_KEY"] = final_df["SCHOOL_ID"].astype(str).str[:5].str.lstrip("0")
absence_panel["DIST_KEY"] = absence_panel["DISTRICT_ID"].astype(str).str[:6].str.lstrip("0")

print("Panel DIST_KEY:", final_df["DIST_KEY"].head(5).tolist())
print("Absence DIST_KEY:", absence_panel["DIST_KEY"].head(5).tolist())
```

```python
ml_df = pd.merge(final_df, absence_panel[["DIST_KEY", "YEAR", "CHRONIC_ABSENT_RATE"]],
                 on=["DIST_KEY", "YEAR"], how="left")

print(f"Shape: {ml_df.shape}")
print(f"Missing target: {ml_df['CHRONIC_ABSENT_RATE'].isna().sum()}")
print(ml_df[["SCHOOL_ID", "YEAR", "FRL_PERCENT", "CHRONIC_ABSENT_RATE"]].head(10))
```

```python
ml_df_clean = ml_df.dropna(subset=["CHRONIC_ABSENT_RATE"])

print(f"Final ML dataset shape: {ml_df_clean.shape}")
print(f"Unique schools: {ml_df_clean['SCHOOL_ID'].nunique()}")
print(f"Years: {ml_df_clean['YEAR'].unique()}")

ml_df_clean.to_csv("/content/final_ML_dataset.csv", index=False)
print("Saved!")
ml_df_clean.head()
```
